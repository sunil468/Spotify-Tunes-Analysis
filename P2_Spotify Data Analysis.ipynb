### Top Ten Song Having Popularity >= 90

```python
most_popular_song = df_track.query('popularity>=90', inplace = False).sort_values('popularity', ascending = False)#.head(10)
#  "inplace = False" means we don't change the dataframe
most_popular_song[:10]
```

### Change The Index As Release date

```python
df_track.set_index('release_date', inplace = True)
df_track.index = pd.to_datetime(df_track.index)
df_track.head()
```

### Artist Present In The 18th Index Of Dataset

```python
df_track[['artists']].iloc[18]
```

### Convert The Duration From Millisecond To Second

```python
df_track['duration'] = df_track['duration_ms'].apply(lambda x: round(x/1000))
df_track.drop('duration_ms', inplace = True, axis = 1)
df_track.duration.head()
```

### Visualize The Correlation Between All The Attribute

```python
corr_df = df_track.drop(['key','mode','explicit'],axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', linewidths = 1, linecolor = 'Black')
#annot - write data value in each cell , fmt - string format used in annotation, cmap - colormap
heatmap.set_title("Corelation Between All The Attribute")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

### Creating a Sample Dataframe for different observation

```python
sample_df = df_track.sample(int(0.004*len(df_track)))
print(sample_df)
```

### Regression Plot Between Loudness & Energy

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = 'Loudness vs Energy Correlation')
```

### Regression Plot Between Popularity & Acousticness

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'b').set(title = 'popularity vs acousticness Correlation')
```

### Creation Of Year Coloumn From Release Date Coloumn

```python
df_track['dates'] = df_track.index.get_level_values('release_date')
df_track.dates = pd.to_datetime(df_track.dates)
years = df_track.dates.dt.year 
```

```python
### Distributive Plot To Visualize Total Number Of Songs In Each Year
sns.displot(years,discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number Of Songs Per Year')
```

### Duration Of Songs Over Year

```python
total_dr = df_track.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = 'Year vs Duration')
plt.xticks(rotation = 90)
```

### Line Plot: Average Durations Of Songs Over Years

```python
total_dr = df_track.duration
sns.set_style(style = 'whitegrid')
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = 'Year vs Duration')
plt.xticks(rotation = 60)
```
